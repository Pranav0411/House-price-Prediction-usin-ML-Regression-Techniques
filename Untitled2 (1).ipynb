```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import sklearn
```

```python
from sklearn.datasets import load_boston
```

```python
bos=load_boston()
```

```python
df=pd.DataFrame(bos.data , columns=bos.feature_names)
```

```python
x=df
```

```python
y=bos.target
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
f=smf.ols(formula='y ~ CRIM + ZN + CHAS + NOX',data=df)
model=f.fit()
model.summary()
```

```python
xc=sm.add_constant(x)
```

```python
pd.DataFrame(xc)
```

```python
model=sm.OLS(y,xc)
```

```python
lr=model.fit()
```

```python
lr.summary()
```
